```python
from mtcnn import MTCNN
from PIL import Image
import cv2
from matplotlib.pyplot import imshow
```

```python
path = "dataset/taylorswift_1.jpg"
```

```python
#mcnn parser
image = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
detector = MTCNN()
result = detector.detect_faces(image)
bounding_box = result[0]['box']
keypoints = result[0]['keypoints']
```

```python
pil_im = Image.open(path)

crop_im = pil_im.crop((bounding_box[0], bounding_box[1],bounding_box[0]+ bounding_box[2], bounding_box[1] + bounding_box[3] ))
imshow(crop_im)
# crop_im.show()
```

```python
left = bounding_box[0]
upper = bounding_box[1]
width = bounding_box[2]
height = bounding_box[3]

crop_im = pil_im.crop((left , upper, 1.60*width + left, 1.60*height + upper))
imshow(crop_im)
```
